```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'C:\Users\anirv\Downloads\new_retail_data.csv\new_retail_data.csv')

print(f"Data Loaded: {data.shape[0]} rows, {data.shape[1]} columns")
```

```python
datab = data.copy()
data.columns.str.strip()
```

```python
data = data.drop(columns=['Transaction_ID','Name','Phone','Email'], errors='ignore')
```

```python
data.info()
```

```python
data['Customer_ID'] = data['Customer_ID'].astype(str).str.replace('.0','', regex=False)
data['Zipcode'] = data['Zipcode'].astype(str).str.replace('.0','', regex=False)
```

```python
num_cols = ['Age','Total_Purchases','Amount','Total_Amount','Ratings']

for col in num_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')
       
```

```python
data['Income'] = data['Income'].str.strip().str.lower().astype('category')
```

```python
data['Income'] = data['Income'].str.strip().str.lower().astype('category')
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['Year'] = pd.to_numeric(data['Year'], errors='coerce').astype('Int64')
```

```python
cat_cols = ['City','State','Country','Gender','Customer_Segment',
            'Product_Category','Product_Brand','Product_Type',
            'Feedback','Shipping_Method','Payment_Method','Order_Status']

for col in cat_cols:
    data[col] = data[col].astype('category')
```

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
num_cols = ['Total_Purchases','Total_Amount','Amount','Ratings','Age']

for col in num_cols:
    data[col] = data[col].fillna(data[col].median())
```

```python
data = data.dropna(subset=['Date'])
data = data.drop(columns=['Time'], errors='ignore')
```

```python
cat_cols = ['Order_Status','Gender','Product_Category','Product_Brand',
            'Feedback','Payment_Method','Shipping_Method',
            'Customer_Segment','Product_Type','Income']

for col in cat_cols:
    data[col] = data[col].astype('object').fillna('unknown').astype('category')
```

```python
data = data.drop(columns = ['Month','Year'],errors = 'ignore')
```

```python
data['Month_Name'] = data['Date'].dt.month_name()
data['Year'] = data['Date'].dt.year
data['Address'] = data['Address'].fillna('unknown')
```

```python
data = data.drop_duplicates()
```
